```python
# SQLAlchemy
import pandas as pd
```

```python
attendance_file = "Resources/Attendance.csv"
att_data_df = pd.read_csv(attendance_file)
att_data_df.head(15)
```

```python
team_file = "Resources/Team_valuations.csv"
team_data_df = pd.read_csv(team_file)
team_data_df.head()
```

```python
#merge the DataFrames to one uusing inner join
merged_df = pd.merge(att_data_df,team_data_df, on = "TEAM" )
merged_df.dtypes
```

```python
#use "{:,}"  to large numbers for easy read 
merged_df["TOTAL"] = merged_df["TOTAL"].map("{:,}".format)
merged_df["AVG"] = merged_df["AVG"].map("{:,}".format)
merged_df.head()
```

```python
#drop GMs column
del merged_df['GMS']
merged_df

```

```python
#add team ticker column
team_ticker = ['CB','DM','SK','MH','TR','LAC','CC','GSW','OCT','NYK','BS','LAL','PTB','SAS','UJ','OM','NOP','HR','PS','IP','MG','CH','76ERS','AH','BN','MB','WW','DN','MT','DP']
merged_df['TEAM TICKER']=team_ticker
merged_df
```

```python
newteam_data=merged_df.set_index('TEAM')
newteam_data.head()
```

```python
newteam_data.sort_values(["VALUE_MILLIONS"], ascending = False)
```

```python
#import our necessary libraries 
from sqlalchemy import create_engine
```

```python
## Create Database Connection and ensure that our table exists with create_all()
engine = create_engine('sqlite:///NBA_Social.sqlite', echo=True)

sqlite_connection = engine.connect()


```

```python
sqlite_table = "NBAteams"
newteam_data.to_sql(sqlite_table, sqlite_connection, if_exists='fail')
```

```python
sqlite_connection.close()
```

```python
from sqlalchemy.ext.automap import automap_base
```

```python
# Create engine using the `demographics.sqlite` database file
engine = create_engine('sqlite:///NBA_Social.sqlite')
```

```python
# Declare a Base using `automap_base()`
Base = automap_base()
```

```python
# Use the Base class to reflect the database tables
Base.prepare(engine, reflect=True)
```

```python
# Print all of the classes mapped to the Base
Base.classes.keys()
```

```python
#Query all records

data = engine.execute ('SELECT * FROM NBAteams')

for record in data:
    print(record)
```
